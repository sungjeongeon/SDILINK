```python
# Missing values:

def missingValue(df):
    #Identifying Missing data.
    total_null = df.isnull().sum().sort_values(ascending = False)
    percent = ((df.isnull().sum()/len(df))*100).sort_values(ascending = False)
    print(f"Total records in our data =  {df.shape[0]} where missing values are as follows:")

    missing_data = pd.concat([total_null,percent.round(2)],axis=1,keys=['Total Missing','In Percent'])
    return missing_data
```

```python
missing_df = missingValue(fuel_cells_df)
missing_df[missing_df['Total Missing'] > 0]
```

```python
fuel_cells_df['datetime'] = fuel_cells_df['datetime'].astype('int64')
fuel_cells_df['datetime'].head()
```

```python
# copy before mormalization and after datetime converson to int
fuel_cells_df_copy2 = fuel_cells_df.copy()
```

```python
cols = ['cycle', 'datetime', 'capacity', 'voltage_measured', 'current_measured',
       'temperature_measured', 'current_load', 'voltage_load', 'time', 'flag']
mm_scaler = MinMaxScaler()

fuel_cells_df[cols] = mm_scaler.fit_transform(fuel_cells_df[cols])
fuel_cells_df.head()
```

```python
fuel_cells_df['flag'].unique()
```

```python
from sklearn.ensemble import IsolationForest
```

```python
flag_df = pd.DataFrame()
flag_df_final = pd.DataFrame()
```

```python
fuel_cells_df
```

```python
contamination_value = 0.01

# Create a copy of the original DataFrame to avoid modifying it
fuel_cells_df_final = fuel_cells_df.copy()

# Create a variable holding the relevant columns for fitting the model and decision_function
feature_columns = fuel_cells_df.columns.difference(['flag','time','datetime'])
```

```python
feature_columns
```

```python
# Initialize the IsolationForest model with the desired contamination value
model = IsolationForest(n_estimators=200, max_samples='auto', contamination=contamination_value, random_state=42)
```

```python
# Fit the model with the feature columns
model.fit(fuel_cells_df[feature_columns])

# Predict the anomaly scores and add them to the DataFrame
predictions = model.predict(fuel_cells_df[feature_columns])
fuel_cells_df_final['anomaly_score' + '_' + str(contamination_value)] = predictions

# Compute the decision function scores and add them to the DataFrame
scores = model.decision_function(fuel_cells_df[feature_columns])
fuel_cells_df_final['scores' + '_' + str(contamination_value)] = scores
```

```python
fuel_cells_df_final
```

```python
fuel_cells_df_final['anomaly_score_0.01'].value_counts(ascending= True)
```

```python
# Save the trained model to a file
with open('isolation_forest_model-0.1.0.pkl', 'wb') as file:
    pickle.dump(model, file)
```
